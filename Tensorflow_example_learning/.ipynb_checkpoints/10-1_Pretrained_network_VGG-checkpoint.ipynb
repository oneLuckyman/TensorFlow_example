```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np 
import glob 
import os
```

```python
train_image_path = glob.glob('../../Tensorflow_course/dataset/all_dataset/dataset/dc_2000/train/*/*.jpg')
```

```python
len(train_image_path)
```

```python
train_image_label = [int(p.split('\\')[1] == 'cat') for p in train_image_path]
```

```python
train_image_label[-5:]
```

```python
train_image_label[:5]
```

```python
def load_preprocess_image(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [256, 256])
    image = tf.cast(image, tf.float32)
    image = image/255
    return image, label
```

```python
train_image_ds = tf.data.Dataset.from_tensor_slices((train_image_path, train_image_label))
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE
```

```python
train_image_ds = train_image_ds.map(load_preprocess_image, num_parallel_calls=AUTOTUNE)
```

```python
train_image_ds
```

```python
for img, label in train_image_ds.take(2):
    plt.imshow(img)
```

```python
BATCH_SIZE = 32 
train_count = len(train_image_path)
```

```python
train_image_ds = train_image_ds.shuffle(train_count).repeat().batch(BATCH_SIZE)
```

```python
test_image_path = glob.glob('../../Tensorflow_course/dataset/all_dataset/dataset/dc_2000/test/*/*.jpg')
test_image_label = [int(p.split('\\')[1] == 'cat') for p in test_image_path]
test_image_ds = tf.data.Dataset.from_tensor_slices((test_image_path, test_image_label))
test_image_ds = test_image_ds.map(load_preprocess_image, num_parallel_calls=AUTOTUNE)
test_image_ds = test_image_ds.repeat().batch(BATCH_SIZE)
```

```python
test_count = len(test_image_path)
test_count
```

# 

```python
conv_base = tf.keras.applications.VGG16(weights='imagenet', include_top=False)
```

```python
conv_base.summary()
```

```python
model = tf.keras.Sequential()
model.add(conv_base)
model.add(tf.keras.layers.GlobalAveragePooling2D())
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
conv_base.trainable = False
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001),
             loss='binary_crossentropy',
             metrics=['acc'])
```
